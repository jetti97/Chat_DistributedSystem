````python
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL, use_fast=False)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

def load_jsonl_as_dataset(path):
    ds = load_dataset("json", data_files=path, split="train")
    # Check columns
    print("Columns:", ds.column_names)
    assert "prompt" in ds.column_names and "target" in ds.column_names, "❌ JSONL braucht Spalten: prompt, target"
    return ds

det_ds = load_jsonl_as_dataset(TRAIN_DET)
slot_ds = load_jsonl_as_dataset(TRAIN_SLOT)

print("DET example:", det_ds[0])
print("SLOT example:", slot_ds[0])
````

````python
MAX_LEN = 512  # kannst du auf 768/1024 erhöhen, wenn deine Slot-Prompts lang sind

def tokenize_and_mask(example):
    prompt = example["prompt"]
    target = example["target"]

    # full text (prompt + target)
    full_text = prompt + target + tokenizer.eos_token

    # tokenize prompt alone (für mask)
    prompt_ids = tokenizer(prompt, add_special_tokens=False).input_ids
    full = tokenizer(
        full_text,
        truncation=True,
        max_length=MAX_LEN,
        padding=False,
        add_special_tokens=False
    )

    labels = full["input_ids"].copy()

    # mask prompt tokens
    cut = min(len(prompt_ids), len(labels))
    for i in range(cut):
        labels[i] = -100

    full["labels"] = labels
    return full

det_tok = det_ds.map(tokenize_and_mask, remove_columns=det_ds.column_names)
slot_tok = slot_ds.map(tokenize_and_mask, remove_columns=slot_ds.column_names)

print("✅ Tokenized")
print("DET token sample keys:", det_tok[0].keys())
````

````python
def build_lora_model():
    model = AutoModelForCausalLM.from_pretrained(
        BASE_MODEL,
        torch_dtype=torch.bfloat16, # Changed from torch.float16 to torch.bfloat16
        device_map="auto"
    )
    model.config.use_cache = False  # wichtig für Training
    model.gradient_checkpointing_enable()

    lora = LoraConfig(
        task_type=TaskType.CAUSAL_LM,
        r=16,
        lora_alpha=32,
        lora_dropout=0.05,
        bias="none",
        target_modules=["q_proj","k_proj","v_proj","o_proj","gate_proj","up_proj","down_proj"]
    )

    model = get_peft_model(model, lora)
    model.print_trainable_parameters()
    return model
````

````python
def train_lora(tokenized_ds, out_dir, run_name, epochs=2, lr=2e-4, batch_size=1, grad_acc=16):
    model = build_lora_model()

    args = TrainingArguments(
        output_dir=out_dir,
        per_device_train_batch_size=batch_size,
        gradient_accumulation_steps=grad_acc,
        learning_rate=lr,
        num_train_epochs=epochs,
        logging_steps=10,
        save_strategy="epoch",
        save_total_limit=2,
        bf16=True, # Changed from fp16=True to bf16=True
        report_to="none",
        optim="adamw_torch",
        warmup_ratio=0.05,
        lr_scheduler_type="cosine",
        run_name=run_name
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=tokenized_ds,
        tokenizer=tokenizer
    )

    trainer.train()

    model.save_pretrained(out_dir)
    tokenizer.save_pretrained(out_dir)
    print(f"✅ Saved LoRA to: {out_dir}")
````

## Parameter für die Textgenerierung und ihre Auswirkung auf das Ergebnis

Beim Generieren von Text mit großen Sprachmodellen können verschiedene Parameter eingestellt werden, um das Verhalten des Modells zu steuern. Diese beeinflussen die Qualität, Länge und den Stil des generierten Textes erheblich.

### 1. `max_new_tokens`
*   **Beschreibung**: Definiert die maximale Anzahl von *neuen* Tokens, die das Modell nach dem Prompt generieren soll. Die Generation stoppt, sobald diese Anzahl erreicht ist oder ein End-of-Sequence (EOS)-Token generiert wird.
*   **Auswirkung**: Direkte Kontrolle über die Länge der Antwort. Ein zu kleiner Wert kann dazu führen, dass Antworten abgeschnitten werden; ein zu großer Wert kann zu unnötig langen oder abschweifenden Texten führen. Für unsere Detektionsaufgabe, die nur ein Token ("JA" oder "NEIN") erfordert, ist ein kleiner Wert wie `5` sinnvoll.

### 2. `pad_token_id` (und `eos_token_id`)
*   **Beschreibung**: `pad_token_id` ist der Token, der verwendet wird, um kürzere Eingabesequenzen auf die maximale Länge einer Batch-Eingabe aufzufüllen. `eos_token_id` ist der Token, der das Ende einer generierten Sequenz signalisiert. Viele Modelle verwenden den `eos_token_id` auch als `pad_token_id`.
*   **Auswirkung**: `pad_token_id` ist wichtig für das Batching von Eingaben während der Generation. Das Modell stoppt die Generation normalerweise, wenn es seinen `eos_token_id` generiert. Ohne einen definierten `pad_token` kann es zu Fehlern beim Batching oder bei der Generierung kommen.

### 3. `do_sample`
*   **Beschreibung**: Ein Boolescher Wert, der bestimmt, ob die Textgenerierung auf **Sampling** (zufällige Auswahl basierend auf Wahrscheinlichkeiten) oder **Greedy Decoding** (immer das wahrscheinlichste nächste Token wählen) basiert.
    *   `do_sample=False`: Greedy Decoding (deterministisch, kann zu repetitiven oder generischen Antworten führen).
    *   `do_sample=True`: Sampling (führt zu vielfältigeren und kreativeren Antworten, aber auch zu mehr Zufälligkeit).
*   **Auswirkung**: Bei `do_sample=True` werden weitere Parameter wie `temperature`, `top_k` und `top_p` aktiv, um das Sampling zu steuern. Wenn `do_sample=False`, werden diese Parameter ignoriert.

### 4. `temperature`
*   **Beschreibung**: Wird nur bei `do_sample=True` verwendet. Ein Wert größer als 0, der die Zufälligkeit der generierten Ausgabe steuert. Kleinere Werte (`< 1.0`) machen die Modellvorhersagen deterministischer, indem sie die Wahrscheinlichkeitsverteilung zugunsten der wahrscheinlicheren Tokens schärfen. Größere Werte (`> 1.0`) erhöhen die Zufälligkeit, indem sie die Wahrscheinlichkeitsverteilung abflachen.
*   **Auswirkung**: Niedrige `temperature` führt zu konservativeren, oft präziseren und weniger überraschenden Texten. Hohe `temperature` fördert kreative, aber potenziell inkonsistente oder unsinnige Texte.

### 5. `top_k`
*   **Beschreibung**: Wird nur bei `do_sample=True` verwendet. Der Modell wählt das nächste Wort nur aus den `k` wahrscheinlichsten Wörtern aus.
*   **Auswirkung**: Begrenzt den Wortschatz für das Sampling auf die top-k wahrscheinlichsten Tokens. Dies reduziert die Wahrscheinlichkeit, dass sehr unwahrscheinliche (und oft unsinnige) Tokens generiert werden, kann aber auch die Vielfalt einschränken.

### 6. `top_p` (Nucleus Sampling)
*   **Beschreibung**: Wird nur bei `do_sample=True` verwendet. Das Modell wählt das nächste Wort aus der kleinsten Menge von Wörtern aus, deren kumulierte Wahrscheinlichkeit `p` überschreitet.
*   **Auswirkung**: Bietet eine dynamischere Kontrolle über die Vielfalt als `top_k`. Es passt sich der Verteilung der Wahrscheinlichkeiten an: Wenn die Verteilung spitz ist (einige wenige Tokens sind sehr wahrscheinlich), wählt `top_p` aus einer kleinen Menge; wenn die Verteilung flacher ist, wählt es aus einer größeren Menge. Dies kann zu natürlicheren und vielfältigeren Texten führen als ein fester `top_k` Wert.

### 7. `num_beams` (für Beam Search)
*   **Beschreibung**: Wird nur bei `do_sample=False` verwendet. Aktiviert die Beam Search-Generierungsstrategie. Anstatt nur den wahrscheinlichsten Pfad zu verfolgen (wie bei Greedy Decoding), hält Beam Search `num_beams` viele vielversprechende Kandidatensequenzen gleichzeitig im Auge.
*   **Auswirkung**: Führt zu qualitativ besseren, kohärenteren und grammatikalisch korrekteren Texten, da das Modell einen breiteren Suchraum erkundet, um die optimale Sequenz zu finden. Allerdings ist es rechenintensiver und langsamer als Greedy Decoding. Es ist deterministisch, ähnlich wie Greedy Decoding, aber 'intelligenter' in seiner Auswahl.


**Zusammenfassend:** Die Wahl der Generationparameter hängt stark von der gewünschten Ausgabe und der Art der Aufgabe ab. Für Aufgaben, die präzise und kurze Antworten erfordern (wie JA/NEIN-Detektion), sind oft deterministische Ansätze oder Sampling mit sehr geringer `temperature` und kontrolliertem `max_new_tokens` bevorzugt. Für kreative oder offene Generierungen sind höhere `temperature`, `top_k` und `top_p` in Kombination mit `do_sample=True` besser geeignet.

## Inference mit den trainierten LoRA-Adaptern

Um die LoRA-Adapter für die Inferenz zu nutzen, müssen wir das Basismodell laden, dann die LoRA-Adapter hinzufügen und sie optional zusammenführen, um ein vollwertiges feingetuntes Modell für die Textgenerierung zu erhalten.

````python
# Bibliotheken importieren
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel

# Die im Training verwendeten Konstanten
BASE_MODEL = "mistralai/Mistral-7B-Instruct-v0.2"
OUT_DET  = "/content/lora_det_final"
OUT_SLOT = "/content/lora_slot_final"

# Tokenizer laden (der selbe, der im Training verwendet wurde)
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL, use_fast=False)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

print("✅ Tokenizer geladen.")
````

### LoRA-Adapter für Detektion laden und mergen

````python
# Basismodell laden
base_model_det = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

# LoRA-Adapter für Detektion laden
lora_model_det = PeftModel.from_pretrained(base_model_det, OUT_DET)

# Adapter mit dem Basismodell zusammenführen (optional, aber gut für die Inferenz)
# Dies erstellt ein 'vollständiges' feingetuntes Modell, das nicht mehr auf das Basismodell angewiesen ist.
merged_model_det = lora_model_det.merge_and_unload()

print(f"✅ LoRA-Adapter für Detektion von {OUT_DET} geladen und zusammengeführt.")
print(f"Modell für Detektion auf Gerät: {merged_model_det.device}")
````

### Inferenz mit dem Detektionsmodell

````python
def generate_response(model, tokenizer, prompt, max_new_tokens=50):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=max_new_tokens, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Debug print to see full response
    print(f"Full generated response (for debug): {response}")

    # Nur den generierten Teil nach dem Prompt zurückgeben
    # Dies erfordert möglicherweise eine Anpassung, je nachdem wie der Prompt strukturiert ist
    return response[len(prompt):].strip()

# Beispiel-Prompt für Detektion
det_prompt = "Du bist Requirements Engineer. Entscheide, ob der folgende Use-Case-Schritt mindestens eine normative Systemanforderung enthält. Antworte ausschließlich mit genau einem Token: JA oder NEIN.\n\nUC_ID: A-01-01_ Login\nSchritt_ID: 1\nText: Der Benutzer öffnet die Login-Seite.\n"

print("--- Detektion Inferenz ---")
print(f"Prompt: {det_prompt}")
det_output = generate_response(merged_model_det, tokenizer, det_prompt, max_new_tokens=5) # Reduced max_new_tokens
print(f"Generierte Antwort: {det_output}")
````

### LoRA-Adapter für Slot-Extraktion laden und mergen

````python
# Basismodell erneut laden (oder das vorherige zurücksetzen, wenn es dasselbe ist)
# Für dieses Beispiel laden wir es neu, um Klarheit zu schaffen.
base_model_slot = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

# LoRA-Adapter für Slot-Extraktion laden
lora_model_slot = PeftModel.from_pretrained(base_model_slot, OUT_SLOT)

# Adapter mit dem Basismodell zusammenführen
merged_model_slot = lora_model_slot.merge_and_unload()

print(f"✅ LoRA-Adapter für Slot-Extraktion von {OUT_SLOT} geladen und zusammengeführt.")
print(f"Modell für Slot-Extraktion auf Gerät: {merged_model_slot.device}")
````

````python
import json
import torch
import pandas as pd
from pathlib import Path
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel

BASE_MODEL = "mistralai/Mistral-7B-Instruct-v0.2"
TEST_STEPS_XLSX = "/content/rover_steps_all_TEST.xlsx"

assert Path(TEST_STEPS_XLSX).exists(), "❌ rover_steps_all_TEST.xlsx fehlt"
assert Path(DET_ADAPTER_DIR).exists(), "❌ DET adapter fehlt"
assert Path(SLOT_ADAPTER_DIR).exists(), "❌ SLOT adapter fehlt"
assert torch.cuda.is_available(), "❌ GPU fehlt"

print("✅ GPU:", torch.cuda.get_device_name(0))

# 4bit quant
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True
)

tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

def load_model(adapter_dir: Path):
    base = AutoModelForCausalLM.from_pretrained(
        BASE_MODEL,
        quantization_config=bnb_config,
        device_map="auto",
        torch_dtype=torch.float16
    )
    model = PeftModel.from_pretrained(base, str(adapter_dir))
    model.eval()
    return model

print("Loading DET model...")
det_model = load_model(Path(DET_ADAPTER_DIR))

print("Loading SLOT model...")
slot_model = load_model(Path(SLOT_ADAPTER_DIR))

# -------------------------
# FEW-SHOT (wie T3) - minimal, aber identisch vom Prinzip:
# - DET: JA/NEIN
# - SLOT: JSON P/A/R/I/S
# -------------------------
FEWSHOT_DET = """Beispiel 1:
Text: "Rover dokumentiert die Aktivität"
Antwort: {"is_requirement": true}

Beispiel 2:
Text: "Rover empfängt die Mission von MCC"
Antwort: {"is_requirement": false}
"""

FEWSHOT_SLOT = """Beispiel 1:
Text: "Rover dokumentiert die Aktivität"
Antwort: [{"P":"-","A":"Rover","R":"-","I":"dokumentiert die Aktivität","S":"-"}]
"""

def det_prompt(text: str):
    return (
        "Du bist Expert:in für Requirements Engineering. "
        "Entscheide, ob der Text mindestens eine Systemanforderung enthält.\n"
        "Antworte NUR als JSON: {\"is_requirement\": true/false}\n\n"
        + FEWSHOT_DET +
        f"\nText: \"{text}\"\nAntwort:"
    )

def slot_prompt(text: str):
    return (
        "Du bist Expert:in für Requirements Engineering. "
        "Extrahiere alle Systemanforderungen als PARIS-Slots.\n"
        "Antworte NUR als JSON-Array von Objekten mit Keys P,A,R,I,S.\n\n"
        + FEWSHOT_SLOT +
        f"\nText: \"{text}\"\nAntwort:"
    )

def generate(model, prompt, max_new_tokens=200):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=False
        )
    return tokenizer.decode(out[0], skip_special_tokens=True)

def safe_json_obj(text: str):
    s = text.find("{")
    e = text.rfind("}")
    if s == -1 or e == -1:
        return None
    try:
        return json.loads(text[s:e+1])
    except:
        return None

def safe_json_array(text: str):
    s = text.find("[")
    e = text.rfind("]")
    if s == -1 or e == -1:
        return []
    try:
        obj = json.loads(text[s:e+1])
        return obj if isinstance(obj, list) else []
    except:
        return []

test = pd.read_excel(TEST_STEPS_XLSX)
test = test[test["Originaltext"].notna()].copy()

det_rows = []
slot_rows = []

print("🚀 Running inference on rows:", len(test))

for i, row in test.iterrows():
    uc = str(row["UC_ID"])
    sid = str(row["Schritt_ID"])
    txt = str(row["Originaltext"]).strip()

    # Detection
    det_out = generate(det_model, det_prompt(txt), max_new_tokens=80)
    det_js = safe_json_obj(det_out) or {"is_requirement": False}
    is_req = bool(det_js.get("is_requirement", False))

    det_rows.append({"UC_ID": uc, "Schritt_ID": sid, "y_pred": is_req})

    # Slotting nur wenn requirement
    if is_req:
        slot_out = generate(slot_model, slot_prompt(txt), max_new_tokens=220)
        arr = safe_json_array(slot_out)
        # nimm "best" = erstes element, falls mehrere
        if arr and isinstance(arr[0], dict):
            best = arr[0]
        else:
            best = {"P":"", "A":"", "R":"", "I":"", "S":""}

        slot_rows.append({
            "UC_ID": uc,
            "Schritt_ID": sid,
            "P": str(best.get("P","")),
            "A": str(best.get("A","")),
            "R": str(best.get("R","")),
            "I": str(best.get("I","")),
            "S": str(best.get("S",""))
        })

det_df = pd.DataFrame(det_rows)
slot_df = pd.DataFrame(slot_rows)

det_path = "/content/preds_detection.csv"
slot_path = "/content/preds_slots_best.csv"

det_df.to_csv(det_path, index=False)
slot_df.to_csv(slot_path, index=False)

print("✅ DONE")
print("Detection rows:", len(det_df))
print("Slot rows:", len(slot_df))
print("Wrote:", det_path, slot_path)
````

### `temperature` – Die Steuerung der Zufälligkeit

`temperature` ist ein Parameter, der die **Wahrscheinlichkeitsverteilung** der möglichen nächsten Token steuert. Stellen Sie sich vor, das Modell schlägt für jedes nächste Wort eine Liste von Kandidaten vor, jeder mit einer bestimmten Wahrscheinlichkeit (z.B. "Katze" mit 80%, "Hund" mit 15%, "Haus" mit 4%, "Baum" mit 1%).

*   **`temperature` = 1.0 (Standard)**: Bei einer Temperatur von 1.0 werden die Wahrscheinlichkeiten so verwendet, wie das Modell sie berechnet hat. Die Wahl des nächsten Tokens erfolgt proportional zu diesen Wahrscheinlichkeiten.

*   **`temperature` < 1.0 (z.B. 0.7)**: Eine niedrigere Temperatur **schärft die Wahrscheinlichkeitsverteilung**. Das bedeutet, dass die Wahrscheinlichkeiten für die bereits wahrscheinlichen Tokens noch höher werden, während die unwahrscheinlicheren Tokens noch unwahrscheinlicher werden. Das Ergebnis ist eine **deterministischere und konservativere** Generierung. Das Modell wählt eher die "sicheren", häufigsten oder vorhersehbarsten Wörter. Dies ist nützlich für Aufgaben, die präzise, faktenbasierte oder weniger kreative Antworten erfordern.

*   **`temperature` > 1.0 (z.B. 1.2)**: Eine höhere Temperatur **glättet die Wahrscheinlichkeitsverteilung**. Das führt dazu, dass die Wahrscheinlichkeiten zwischen allen möglichen Tokens "gerechter" verteilt werden – die Differenzen zwischen wahrscheinlichen und unwahrscheinlichen Tokens werden kleiner. Dies fördert die **Zufälligkeit und Vielfalt** in der Generierung. Das Modell kann "überraschendere" oder kreativere Wörter wählen, birgt aber auch das Risiko, weniger kohärenten oder sogar unsinnigen Text zu erzeugen.

**Analogie**: Denken Sie an ein Roulette-Rad. Eine niedrige Temperatur macht die Sektoren der wahrscheinlichsten Zahlen größer, während eine hohe Temperatur die Sektoren aller Zahlen ähnlicher groß macht, sodass auch die unwahrscheinlichen Zahlen öfter getroffen werden.

### `top_p` (Nucleus Sampling) – Dynamische Wortschatzauswahl

`top_p`, auch bekannt als **Nucleus Sampling**, bietet eine dynamischere und oft effektivere Methode zur Steuerung der Vielfalt als `top_k`. Anstatt eine feste Anzahl von `k` Tokens zu wählen, berücksichtigt `top_p` eine variable Anzahl von Tokens basierend auf deren kumulierten Wahrscheinlichkeit:

*   **Mechanismus**: Bei `top_p=0.9` (ein häufig verwendeter Wert) werden alle Tokens von den höchstwahrscheinlichsten absteigend so lange ausgewählt, bis deren **kumulierte Wahrscheinlichkeit den Schwellenwert `p` überschreitet**. Aus dieser dynamisch erstellten Menge von Tokens wird dann zufällig das nächste Wort ausgewählt.

*   **Vorteile gegenüber `top_k`**:
    *   **Anpassungsfähigkeit**: Wenn die Verteilung der Token-Wahrscheinlichkeiten **spitz** ist (d.h., nur wenige Tokens sind sehr wahrscheinlich), wählt `top_p` automatisch eine **kleine Menge** von Tokens aus. Wenn die Verteilung **flach** ist (d.h., viele Tokens haben ähnliche, aber geringere Wahrscheinlichkeiten), wählt `top_p` eine **größere Menge** von Tokens aus. `top_k` würde in beiden Fällen eine feste `k`-Anzahl wählen, was bei einer spitzen Verteilung zu wenig Vielfalt und bei einer flachen Verteilung zu zu viel Zufälligkeit (und Unsinn) führen könnte.
    *   **Natürlichere Ergebnisse**: Durch diese dynamische Auswahl kann `top_p` oft vielfältigere und gleichzeitig kohärentere Texte erzeugen als `top_k`, da es sich besser an den Kontext und die intrinsische Unsicherheit des Modells anpasst.

**Zusammenwirkung**: `temperature` und `top_p` werden oft in Kombination verwendet (`do_sample=True` muss aktiviert sein). `temperature` kann die grundsätzliche Schärfe oder Glätte der Wahrscheinlichkeitsverteilung vorsteuern, und `top_p` wählt dann auf dieser modifizierten Verteilung dynamisch die Menge der zur Auswahl stehenden Tokens aus. Eine sinnvolle Kombination könnte beispielsweise `temperature=0.7` und `top_p=0.9` sein, um eine Balance zwischen Kreativität und Kohärenz zu finden.

### Modellauswahl und LoRA-Konfiguration

Das **Basismodell** `mistralai/Mistral-7B-Instruct-v0.2` wurde aus folgenden Gründen ausgewählt:
*   **Instruktionsfähigkeit**: Das Modell ist speziell für das Befolgen von Anweisungen trainiert (`Instruct`-Version), was für unsere Aufgaben der Detektion normativer Anforderungen und Extraktion von PARIS-Slots entscheidend ist. Es versteht und verarbeitet strukturierte Prompts effektiv.
*   **Größe**: Mit 7 Milliarden Parametern bietet es eine gute Balance zwischen Leistungsfähigkeit und Recheneffizienz. Es ist groß genug, um komplexe Sprachmuster zu erfassen, aber klein genug, um effizient mit LoRA auf einer einzelnen GPU (z.B. einer A100) feingetunt zu werden.

Die **LoRA-Konfiguration** ist wie folgt definiert:
*   **`r=16` (Rank)**: Dieser Parameter bestimmt den Rang der Update-Matrizen in LoRA. Ein höherer Rang erlaubt mehr Expressivität und eine feinere Anpassung des Modells, erfordert aber auch mehr Speicher und Rechenleistung. `r=16` ist ein guter Ausgangspunkt, der eine effektive Anpassung ermöglicht, ohne zu groß zu sein und Overfitting zu riskieren.
*   **`lora_alpha=32`**: `lora_alpha` ist ein Skalierungsfaktor für die LoRA-Update-Matrizen. Ein höherer Wert gibt den LoRA-Adaptern mehr Gewicht und somit mehr Einfluss auf die Modellaktualisierungen. Das Verhältnis `lora_alpha / r` (hier `32 / 16 = 2`) ist entscheidend; ein Wert > 1 bedeutet, dass die LoRA-Updates stärker gewichtet werden als bei einer rein linearen Skalierung.
*   **`lora_dropout=0.05`**: Dieser Parameter wendet eine Dropout-Rate von 5% auf die LoRA-Schichten an. Dropout ist eine Regularisierungstechnik, die hilft, Overfitting zu verhindern, indem sie zufällig einen Teil der Neuronen während des Trainings deaktiviert. Dies fördert die Robustheit des Modells.
*   **`target_modules`**: Die ausgewählten `target_modules` (`q_proj`, `k_proj`, `v_proj`, `o_proj`, `gate_proj`, `up_proj`, `down_proj`) sind die Linearschichten in den Transformer-Blöcken des Mistral-Modells. Die Auswahl dieser Schichten ist Standard für die Anwendung von LoRA, da sie die wichtigsten Gewichtungsmatrizen für die Aufmerksamkeitsmechanismen (`q_proj`, `k_proj`, `v_proj`, `o_proj`) und die Feed-Forward-Netzwerke (`gate_proj`, `up_proj`, `down_proj`) umfassen. Durch das Anwenden von LoRA auf diese Schlüsselkomponenten kann das Modell effektiv auf die neuen Aufgaben angepasst werden, ohne alle Parameter zu trainieren.

**`torch.bfloat16`** als `torch_dtype` wurde aus folgenden Gründen gewählt:
*   **Speichereffizienz**: `bfloat16` ist ein Gleitkommaformat mit halber Präzision, das weniger Speicherplatz benötigt als `float32`. Dies ist entscheidend für das Training großer Modelle wie Mistral-7B auf GPUs mit begrenztem VRAM.
*   **Rechenleistung**: Moderne GPUs (insbesondere NVIDIA Ampere- und neuere Architekturen wie A100 oder H100) sind für die effiziente Verarbeitung von `bfloat16`-Operationen optimiert. Dies ermöglicht schnellere Trainingszeiten im Vergleich zu `float32`, während die numerische Stabilität (im Vergleich zu `float16`) besser erhalten bleibt.

*   **Spalten**: Jedes JSON-Objekt in der Datei muss die Schlüssel `prompt` und `target` enthalten. Der `prompt` ist die Eingabe für das Modell (z.B. der Use-Case-Schritt mit der Aufgabenstellung), und der `target` ist die erwartete Ausgabe (z.B. JA/NEIN für die Detektion oder das JSON für die Slot-Extraktion).

Nach dem Laden werden die Datensätze `det_ds` und `slot_ds` erstellt, die die Rohdaten für die Detektions- bzw. Slot-Extraktionsaufgaben enthalten.

#### 2. Initialisierung des Tokenizers

Der `AutoTokenizer` wird mit der Methode `AutoTokenizer.from_pretrained(BASE_MODEL, use_fast=False)` initialisiert. Dies lädt den passenden Tokenizer für das angegebene `BASE_MODEL` (hier `mistralai/Mistral-7B-Instruct-v0.2`). Die Option `use_fast=False` stellt sicher, dass der nicht-schnelle (Python-basierte) Tokenizer verwendet wird, was manchmal bei bestimmten Modellen oder Konfigurationen notwendig sein kann.

Ein wichtiger Schritt ist die Überprüfung und Zuweisung des `pad_token`:

```python
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
```

Dies stellt sicher, dass ein Padding-Token verfügbar ist. Viele Modelle, insbesondere wie Mistral, haben standardmäßig keinen expliziten `pad_token` definiert. Für das Training und die Batched-Inferenz ist ein Padding-Token jedoch unerlässlich, um Sequenzen unterschiedlicher Länge auf eine einheitliche `max_length` zu bringen. Durch das Setzen des `pad_token` auf das `eos_token` (End-of-Sentence-Token) wird ein vorhandenes Special Token für diesen Zweck wiederverwendet.

#### 3. Tokenisierung und Maskierung (`tokenize_and_mask` Funktion)

Die Funktion `tokenize_and_mask(example)` verarbeitet ein einzelnes Datenbeispiel und bereitet es für das Training vor:

*   **Extraktion von `prompt` und `target`**: Zuerst werden der `prompt` und `target` aus dem Eingabe-`example` extrahiert.

*   **Erstellung des vollständigen Texts**: Der `full_text` wird durch Aneinanderreihen von `prompt`, `target` und dem `tokenizer.eos_token` erstellt. Der `eos_token` signalisiert dem Modell das Ende der Sequenz und ist wichtig für die Sprachmodellierung.

*   **Tokenisierung des `prompt` und des `full_text`**:
    *   Der `prompt` wird separat tokenisiert (`prompt_ids`), um seine Länge in Token zu bestimmen. Diese Information wird später für die Maskierung benötigt. Es werden keine Special Tokens hinzugefügt (`add_special_tokens=False`).
    *   Der `full_text` wird ebenfalls tokenisiert. Dabei wird `truncation=True` verwendet, um sicherzustellen, dass die Sequenz nicht länger als `MAX_LEN` wird. `MAX_LEN` ist ein Hyperparameter (hier 512), der die maximale Anzahl von Token pro Sequenz festlegt. Längere Sequenzen werden abgeschnitten. `padding=False` bedeutet, dass die Sequenzen zu diesem Zeitpunkt noch nicht mit Padding-Token aufgefüllt werden.

*   **Maskierung der Labels**: Dies ist der entscheidende Schritt für das Fine-Tuning des Modells, um es zu einem instruktionsfolgenden oder generativen Modell zu machen (oft als Causal Language Modeling im Kontext von Fine-Tuning bezeichnet).
    *   Die `labels` werden zunächst als Kopie der `input_ids` des `full_text` erstellt.
    *   Anschließend werden die Token-IDs, die dem `prompt` entsprechen, in der `labels`-Liste auf `-100` gesetzt. Das ist eine Konvention in der Hugging Face `transformers`-Bibliothek: Token mit dem Label `-100` werden bei der Berechnung des Verlusts ignoriert.
    *   **Bedeutung der Maskierung**: Durch diese Maskierung lernt das Modell während des Trainings, den `target`-Teil *basierend auf dem `prompt`* zu generieren, aber der `prompt` selbst trägt nicht zur Verlustberechnung bei. Das Modell wird also dafür trainiert, nur die **Antwort** (den `target`) zu produzieren, nicht die gesamte Eingabe zu wiederholen.

#### 4. Anwendung der Tokenisierung

Die `tokenize_and_mask`-Funktion wird auf die geladenen Datensätze `det_ds` und `slot_ds` mithilfe der `map`-Methode angewendet:

```python
det_tok = det_ds.map(tokenize_and_mask, remove_columns=det_ds.column_names)
slot_tok = slot_ds.map(tokenize_and_mask, remove_columns=slot_ds.column_names)
```

Die `map`-Methode wendet die Funktion auf jedes Element des Datensatzes an. `remove_columns=det_ds.column_names` stellt sicher, dass die ursprünglichen `prompt` und `target` Spalten entfernt werden, da wir nun die `input_ids`, `attention_mask` und `labels` für das Training benötigen.

Das Ergebnis sind die `det_tok` und `slot_tok` Datensätze, die nun die für das Training vorbereiteten tokenisierten und maskierten Daten enthalten.

### Details zur `train_lora`-Funktion und dem Fine-Tuning-Prozess

Die Funktion `train_lora` ist der zentrale Einstiegspunkt für das Fine-Tuning der LoRA-Adapter für beide Modelle (Detektion und Slot-Füllung). Sie orchestriert den gesamten Trainingsablauf, von der Modellinitialisierung bis zum Speichern der trainierten Adapter.

**1. `train_lora`-Funktion und ihre Parameter:**
Die `train_lora`-Funktion nimmt die folgenden Parameter entgegen:
*   `tokenized_ds`: Das vorverarbeitete Dataset, das die tokenisierten Prompts und Targets enthält, die für das Training verwendet werden sollen.
*   `out_dir`: Der Ausgabepfad, in dem die trainierten LoRA-Adapter und der Tokenizer gespeichert werden.
*   `run_name`: Ein Bezeichner für den aktuellen Trainingslauf, nützlich für das Tracking, falls externe Tools wie Weights & Biases integriert wären.
*   `epochs`: Die Anzahl der Trainingsepochen, d.h. wie oft das gesamte Dataset während des Trainings durchlaufen wird.
*   `lr`: Die anfängliche Lernrate, die bestimmt, wie stark die Modellgewichte bei jedem Optimierungsschritt angepasst werden.
*   `batch_size`: Die Anzahl der Beispiele, die pro Trainingsschritt auf einem einzelnen Gerät verarbeitet werden. Hier auf 1 gesetzt, was bedeutet, dass jedes Beispiel einzeln verarbeitet wird.
*   `grad_acc`: Die Gradientenakkumulationsschritte. Dieser Parameter gibt an, wie viele Schritte die Gradienten akkumuliert werden, bevor ein Optimierungsschritt durchgeführt wird. Bei `batch_size=1` und `grad_acc=16` entspricht dies effektiv einer Batch-Größe von 16.

**2. Initialisierung des LoRA-Modells (`build_lora_model()`):**
Innerhalb von `train_lora` wird die `build_lora_model()`-Funktion aufgerufen. Diese Funktion ist verantwortlich für:
*   Das Laden des `BASE_MODEL` (`mistralai/Mistral-7B-Instruct-v0.2`) mithilfe von `AutoModelForCausalLM.from_pretrained()`. Dabei wird `torch.bfloat16` als `torch_dtype` verwendet, um das Modell im `bfloat16`-Format zu laden, was Speicher spart und die Trainingsgeschwindigkeit auf kompatibler Hardware erhöht. `device_map="auto"` sorgt dafür, dass das Modell automatisch auf die verfügbaren GPUs verteilt wird.
*   `model.config.use_cache = False` wird gesetzt, was für das Training wichtig ist, um Probleme mit dem Gradientenfluss zu vermeiden.
*   `model.gradient_checkpointing_enable()` wird aktiviert, um den Speicherverbrauch während des Trainings weiter zu reduzieren, indem die Aktivierungen bestimmter Schichten nicht im Speicher gehalten, sondern bei Bedarf neu berechnet werden.
*   Eine `LoraConfig` wird erstellt, die die Parameter für den LoRA-Adapter definiert (z.B. `r`, `lora_alpha`, `lora_dropout`, `bias`, `target_modules`).
*   Schließlich wird das Basismodell mit der `get_peft_model()`-Funktion um die LoRA-Adapter erweitert. Die Funktion gibt auch die Anzahl der trainierbaren Parameter des modifizierten Modells aus, was nützlich ist, um den Umfang des Fine-Tunings zu verstehen.

**3. `TrainingArguments` zur Konfiguration des Trainingsprozesses:**
Die `TrainingArguments`-Klasse wird verwendet, um alle Hyperparameter und Konfigurationen für den Trainingsprozess festzulegen:
*   `output_dir`: `/content/lora_det_final` oder `/content/lora_slot_final`. Dies ist der Pfad, unter dem alle Trainingsartefakte, einschließlich Modell-Checkpoints, Logs und der endgültigen LoRA-Adapter, gespeichert werden.
*   `per_device_train_batch_size=1`: Die Anzahl der Trainingsbeispiele, die pro GPU in einem einzigen Schritt verarbeitet werden. Ein Wert von 1 wird oft verwendet, wenn der Speicher sehr begrenzt ist oder wenn die effektive Batch-Größe durch Gradientenakkumulation erhöht wird.
*   `gradient_accumulation_steps=16`: Da die `per_device_train_batch_size` 1 ist, werden die Gradienten von 16 einzelnen Schritten gesammelt und dann einmal angewendet, was einer effektiven Batch-Größe von 16 entspricht. Dies hilft, die Stabilität des Trainings zu verbessern und größere effektive Batch-Größen zu simulieren, ohne den GPU-Speicher direkt zu überlasten.
*   `learning_rate=2e-4`: Die anfängliche Lernrate für den Optimierer. Dieser Wert wird im Laufe des Trainings oft durch einen Lernraten-Scheduler angepasst.
*   `num_train_epochs=2`: Das Modell wird zweimal über das gesamte Trainings-Dataset iterieren.
*   `logging_steps=10`: Logs (wie Verlust und Lernrate) werden alle 10 Schritte ausgegeben.
*   `save_strategy="epoch"` und `save_total_limit=2`: Dies bedeutet, dass das Modell am Ende jeder Epoche gespeichert wird, aber nur die zwei neuesten Checkpoints behalten werden, um Speicherplatz zu sparen.
*   `bf16=True`: Aktiviert das `bfloat16`-Mixed-Precision-Training. Dies spart GPU-Speicher und kann die Trainingsgeschwindigkeit erhöhen, insbesondere auf GPUs, die bfloat16-Operationen nativ unterstützen (wie Tensor Cores auf NVIDIA A100 GPUs), ohne die Modellgenauigkeit wesentlich zu beeinträchtigen.
*   `report_to="none"`: Deaktiviert die Integration mit externen Reporting-Tools wie TensorBoard, Weights & Biases oder MLflow.
*   `optim="adamw_torch"`: Der verwendete Optimierungsalgorithmus ist `AdamW` (Adam mit Gewichtsverfall), implementiert von PyTorch. `AdamW` ist eine beliebte Wahl für das Training von Large Language Models.
*   `warmup_ratio=0.05` und `lr_scheduler_type="cosine"`: Dies konfiguriert den Lernraten-Scheduler. Zuerst wird die Lernrate über die ersten 5% der Trainingsschritte linear von 0 auf die angegebene `learning_rate` erhöht (Warmup), und danach folgt sie einem Cosinus-Schema, das die Lernrate allmählich reduziert.
*   `run_name`: Ein eindeutiger Name für den aktuellen Trainingslauf, der in den Logs und bei der Benennung von Verzeichnissen verwendet wird.

**4. Rolle des `Trainer`s:**
Der `Trainer` von Hugging Face ist eine leistungsstarke und benutzerfreundliche API, die den Trainingsprozess abstrahiert. Er wird initialisiert mit:
*   `model`: Das LoRA-Modell, das trainiert werden soll.
*   `args`: Die zuvor definierten `TrainingArguments`.
*   `train_dataset`: Das tokenisierte Trainings-Dataset (`tokenized_ds`).
*   `tokenizer`: Der `AutoTokenizer`, der verwendet wurde, um die Daten zu tokenisieren und für die Decodierung während des Trainings oder der Evaluation notwendig ist.
Der `Trainer` kümmert sich um die Trainingsschleife, Optimierungsschritte, Lernraten-Scheduling, Logging, Checkpoint-Speicherung und vieles mehr.

**5. Starten des Fine-Tunings (`trainer.train()`):**
Durch den Aufruf von `trainer.train()` wird der Fine-Tuning-Prozess gestartet. Der `Trainer` führt die Trainingsschleife gemäß den in `TrainingArguments` definierten Einstellungen aus.

**6. Speichern der LoRA-Adapter und des Tokenizers:**
Nachdem das Training abgeschlossen ist, werden die trainierten LoRA-Adapter und der Tokenizer gespeichert:
*   `model.save_pretrained(out_dir)`: Speichert die trainierbaren LoRA-Adaptergewichte und deren Konfiguration im angegebenen `out_dir`.
*   `tokenizer.save_pretrained(out_dir)`: Speichert die Tokenizer-Konfiguration und das Vokabular im selben `out_dir`. Dies ist entscheidend, um das fein abgestimmte Modell später mit dem korrekten Tokenizer laden und verwenden zu können.

## 3. Datensatzvorbereitung und Tokenisierung

Beschreibung des Ladevorgangs der JSONL-Datensätze, der Tokenisierung unter Verwendung des 'AutoTokenizer' und des Maskierungsprozesses für die Labels.

#### 1. Laden der Datensätze (`load_jsonl_as_dataset`)

Die Funktion `load_jsonl_as_dataset(path)` ist dafür verantwortlich, die Trainingsdaten aus den JSONL-Dateien (`train_det.jsonl` und `train_slot.jsonl`) zu laden. Intern verwendet sie die `load_dataset` Funktion aus der `datasets`-Bibliothek, die das Laden von Daten in verschiedenen Formaten ermöglicht.

*   **Format**: Es wird erwartet, dass die Eingabedateien im JSONL-Format vorliegen, wobei jede Zeile ein JSON-Objekt darstellt.
*   **Spalten**: Jedes JSON-Objekt in der Datei muss die Schlüssel `prompt` und `target` enthalten. Der `prompt` ist die Eingabe für das Modell (z.B. der Use-Case-Schritt mit der Aufgabenstellung), und der `target` ist die erwartete Ausgabe (z.B. JA/NEIN für die Detektion oder das JSON für die Slot-Extraktion).

Nach dem Laden werden die Datensätze `det_ds` und `slot_ds` erstellt, die die Rohdaten für die Detektions- bzw. Slot-Extraktionsaufgaben enthalten.

#### 2. Initialisierung des Tokenizers

Der `AutoTokenizer` wird mit der Methode `AutoTokenizer.from_pretrained(BASE_MODEL, use_fast=False)` initialisiert. Dies lädt den passenden Tokenizer für das angegebene `BASE_MODEL` (hier `mistralai/Mistral-7B-Instruct-v0.2`). Die Option `use_fast=False` stellt sicher, dass der nicht-schnelle (Python-basierte) Tokenizer verwendet wird, was manchmal bei bestimmten Modellen oder Konfigurationen notwendig sein kann.

Ein wichtiger Schritt ist die Überprüfung und Zuweisung des `pad_token`:

```python
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
```

Dies stellt sicher, dass ein Padding-Token verfügbar ist. Viele Modelle, insbesondere wie Mistral, haben standardmäßig keinen expliziten `pad_token` definiert. Für das Training und die Batched-Inferenz ist ein Padding-Token jedoch unerlässlich, um Sequenzen unterschiedlicher Länge auf eine einheitliche `max_length` zu bringen. Durch das Setzen des `pad_token` auf das `eos_token` (End-of-Sentence-Token) wird ein vorhandenes Special Token für diesen Zweck wiederverwendet.

#### 3. Tokenisierung und Maskierung (`tokenize_and_mask` Funktion)

Die Funktion `tokenize_and_mask(example)` verarbeitet ein einzelnes Datenbeispiel und bereitet es für das Training vor:

*   **Extraktion von `prompt` und `target`**: Zuerst werden der `prompt` und `target` aus dem Eingabe-`example` extrahiert.

*   **Erstellung des vollständigen Texts**: Der `full_text` wird durch Aneinanderreihen von `prompt`, `target` und dem `tokenizer.eos_token` erstellt. Der `eos_token` signalisiert dem Modell das Ende der Sequenz und ist wichtig für die Sprachmodellierung.

*   **Tokenisierung des `prompt` und des `full_text`**:
    *   Der `prompt` wird separat tokenisiert (`prompt_ids`), um seine Länge in Token zu bestimmen. Diese Information wird später für die Maskierung benötigt. Es werden keine Special Tokens hinzugefügt (`add_special_tokens=False`).
    *   Der `full_text` wird ebenfalls tokenisiert. Dabei wird `truncation=True` verwendet, um sicherzustellen, dass die Sequenz nicht länger als `MAX_LEN` wird. `MAX_LEN` ist ein Hyperparameter (hier 512), der die maximale Anzahl von Token pro Sequenz festlegt. Längere Sequenzen werden abgeschnitten. `padding=False` bedeutet, dass die Sequenzen zu diesem Zeitpunkt noch nicht mit Padding-Token aufgefüllt werden.

*   **Maskierung der Labels**: Dies ist der entscheidende Schritt für das Fine-Tuning des Modells, um es zu einem instruktionsfolgenden oder generativen Modell zu machen (oft als Causal Language Modeling im Kontext von Fine-Tuning bezeichnet).
    *   Die `labels` werden zunächst als Kopie der `input_ids` des `full_text` erstellt.
    *   Anschließend werden die Token-IDs, die dem `prompt` entsprechen, in der `labels`-Liste auf `-100` gesetzt. Das ist eine Konvention in der Hugging Face `transformers`-Bibliothek: Token mit dem Label `-100` werden bei der Berechnung des Verlusts ignoriert.
    *   **Bedeutung der Maskierung**: Durch diese Maskierung lernt das Modell während des Trainings, den `target`-Teil *basierend auf dem `prompt`* zu generieren, aber der `prompt` selbst trägt nicht zur Verlustberechnung bei. Das Modell wird also dafür trainiert, nur die **Antwort** (den `target`) zu produzieren, nicht die gesamte Eingabe zu wiederholen.

#### 4. Anwendung der Tokenisierung

Die `tokenize_and_mask`-Funktion wird auf die geladenen Datensätze `det_ds` und `slot_ds` mithilfe der `map`-Methode angewendet:

```python
det_tok = det_ds.map(tokenize_and_mask, remove_columns=det_ds.column_names)
slot_tok = slot_ds.map(tokenize_and_mask, remove_columns=slot_ds.column_names)
```

Die `map`-Methode wendet die Funktion auf jedes Element des Datensatzes an. `remove_columns=det_ds.column_names` stellt sicher, dass die ursprünglichen `prompt` und `target` Spalten entfernt werden, da wir nun die `input_ids`, `attention_mask` und `labels` für das Training benötigen.

Das Ergebnis sind die `det_tok` und `slot_tok` Datensätze, die nun die für das Training vorbereiteten tokenisierten und maskierten Daten enthalten.

### Modellauswahl und LoRA-Konfiguration

Das **Basismodell** `mistralai/Mistral-7B-Instruct-v0.2` wurde aus folgenden Gründen ausgewählt:
*   **Instruktionsfähigkeit**: Das Modell ist speziell für das Befolgen von Anweisungen trainiert (`Instruct`-Version), was für unsere Aufgaben der Detektion normativer Anforderungen und Extraktion von PARIS-Slots entscheidend ist. Es versteht und verarbeitet strukturierte Prompts effektiv.
*   **Größe**: Mit 7 Milliarden Parametern bietet es eine gute Balance zwischen Leistungsfähigkeit und Recheneffizienz. Es ist groß genug, um komplexe Sprachmuster zu erfassen, aber klein genug, um effizient mit LoRA auf einer einzelnen GPU (z.B. einer A100) feingetunt zu werden.

Die **LoRA-Konfiguration** ist wie folgt definiert:
*   **`r=16` (Rank)**: Dieser Parameter bestimmt den Rang der Update-Matrizen in LoRA. Ein höherer Rang erlaubt mehr Expressivität und eine feinere Anpassung des Modells, erfordert aber auch mehr Speicher und Rechenleistung. `r=16` ist ein guter Ausgangspunkt, der eine effektive Anpassung ermöglicht, ohne zu groß zu sein und Overfitting zu riskieren.
*   **`lora_alpha=32`**: `lora_alpha` ist ein Skalierungsfaktor für die LoRA-Update-Matrizen. Ein höherer Wert gibt den LoRA-Adaptern mehr Gewicht und somit mehr Einfluss auf die Modellaktualisierungen. Das Verhältnis `lora_alpha / r` (hier `32 / 16 = 2`) ist entscheidend; ein Wert > 1 bedeutet, dass die LoRA-Updates stärker gewichtet werden als bei einer rein linearen Skalierung.
*   **`lora_dropout=0.05`**: Dieser Parameter wendet eine Dropout-Rate von 5% auf die LoRA-Schichten an. Dropout ist eine Regularisierungstechnik, die hilft, Overfitting zu verhindern, indem sie zufällig einen Teil der Neuronen während des Trainings deaktiviert. Dies fördert die Robustheit des Modells.
*   **`target_modules`**: Die ausgewählten `target_modules` (`q_proj`, `k_proj`, `v_proj`, `o_proj`, `gate_proj`, `up_proj`, `down_proj`) sind die Linearschichten in den Transformer-Blöcken des Mistral-Modells. Die Auswahl dieser Schichten ist Standard für die Anwendung von LoRA, da sie die wichtigsten Gewichtungsmatrizen für die Aufmerksamkeitsmechanismen (`q_proj`, `k_proj`, `v_proj`, `o_proj`) und die Feed-Forward-Netzwerke (`gate_proj`, `up_proj`, `down_proj`) umfassen. Durch das Anwenden von LoRA auf diese Schlüsselkomponenten kann das Modell effektiv auf die neuen Aufgaben angepasst werden, ohne alle Parameter zu trainieren.

**`torch.bfloat16`** als `torch_dtype` wurde aus folgenden Gründen gewählt:
*   **Speichereffizienz**: `bfloat16` ist ein Gleitkommaformat mit halber Präzision, das weniger Speicherplatz benötigt als `float32`. Dies ist entscheidend für das Training großer Modelle wie Mistral-7B auf GPUs mit begrenztem VRAM.

### Details zur `train_lora`-Funktion und dem Fine-Tuning-Prozess

Die Funktion `train_lora` ist der zentrale Einstiegspunkt für das Fine-Tuning der LoRA-Adapter für beide Modelle (Detektion und Slot-Füllung). Sie orchestriert den gesamten Trainingsablauf, von der Modellinitialisierung bis zum Speichern der trainierten Adapter.

**1. `train_lora`-Funktion und ihre Parameter:**
Die `train_lora`-Funktion nimmt die folgenden Parameter entgegen:
*   `tokenized_ds`: Das vorverarbeitete Dataset, das die tokenisierten Prompts und Targets enthält, die für das Training verwendet werden sollen.
*   `out_dir`: Der Ausgabepfad, in dem die trainierten LoRA-Adapter und der Tokenizer gespeichert werden.
*   `run_name`: Ein Bezeichner für den aktuellen Trainingslauf, nützlich für das Tracking, falls externe Tools wie Weights & Biases integriert wären.
*   `epochs`: Die Anzahl der Trainingsepochen, d.h. wie oft das gesamte Dataset während des Trainings durchlaufen wird.
*   `lr`: Die anfängliche Lernrate, die bestimmt, wie stark die Modellgewichte bei jedem Optimierungsschritt angepasst werden.
*   `batch_size`: Die Anzahl der Beispiele, die pro Trainingsschritt auf einem einzelnen Gerät verarbeitet werden. Hier auf 1 gesetzt, was bedeutet, dass jedes Beispiel einzeln verarbeitet wird.
*   `grad_acc`: Die Gradientenakkumulationsschritte. Dieser Parameter gibt an, wie viele Schritte die Gradienten akkumuliert werden, bevor ein Optimierungsschritt durchgeführt wird. Bei `batch_size=1` und `grad_acc=16` entspricht dies effektiv einer Batch-Größe von 16.

**2. Initialisierung des LoRA-Modells (`build_lora_model()`):**
Innerhalb von `train_lora` wird die `build_lora_model()`-Funktion aufgerufen. Diese Funktion ist verantwortlich für:
*   Das Laden des `BASE_MODEL` (`mistralai/Mistral-7B-Instruct-v0.2`) mithilfe von `AutoModelForCausalLM.from_pretrained()`. Dabei wird `torch.bfloat16` als `torch_dtype` verwendet, um das Modell im `bfloat16`-Format zu laden, was Speicher spart und die Trainingsgeschwindigkeit auf kompatibler Hardware erhöht. `device_map="auto"` sorgt dafür, dass das Modell automatisch auf die verfügbaren GPUs verteilt wird.
*   `model.config.use_cache = False` wird gesetzt, was für das Training wichtig ist, um Probleme mit dem Gradientenfluss zu vermeiden.
*   `model.gradient_checkpointing_enable()` wird aktiviert, um den Speicherverbrauch während des Trainings weiter zu reduzieren, indem die Aktivierungen bestimmter Schichten nicht im Speicher gehalten, sondern bei Bedarf neu berechnet werden.
*   Eine `LoraConfig` wird erstellt, die die Parameter für den LoRA-Adapter definiert (z.B. `r`, `lora_alpha`, `lora_dropout`, `bias`, `target_modules`).
*   Schließlich wird das Basismodell mit der `get_peft_model()`-Funktion um die LoRA-Adapter erweitert. Die Funktion gibt auch die Anzahl der trainierbaren Parameter des modifizierten Modells aus, was nützlich ist, um den Umfang des Fine-Tunings zu verstehen.

**3. `TrainingArguments` zur Konfiguration des Trainingsprozesses:**
Die `TrainingArguments`-Klasse wird verwendet, um alle Hyperparameter und Konfigurationen für den Trainingsprozess festzulegen:
*   `output_dir`: `/content/lora_det_final` oder `/content/lora_slot_final`. Dies ist der Pfad, unter dem alle Trainingsartefakte, einschließlich Modell-Checkpoints, Logs und der endgültigen LoRA-Adapter, gespeichert werden.
*   `per_device_train_batch_size=1`: Die Anzahl der Trainingsbeispiele, die pro GPU in einem einzigen Schritt verarbeitet werden. Ein Wert von 1 wird oft verwendet, wenn der Speicher sehr begrenzt ist oder wenn die effektive Batch-Größe durch Gradientenakkumulation erhöht wird.
*   `gradient_accumulation_steps=16`: Da die `per_device_train_batch_size` 1 ist, werden die Gradienten von 16 einzelnen Schritten gesammelt und dann einmal angewendet, was einer effektiven Batch-Größe von 16 entspricht. Dies hilft, die Stabilität des Trainings zu verbessern und größere effektive Batch-Größen zu simulieren, ohne den GPU-Speicher direkt zu überlasten.
*   `learning_rate=2e-4`: Die anfängliche Lernrate für den Optimierer. Dieser Wert wird im Laufe des Trainings oft durch einen Lernraten-Scheduler angepasst.
*   `num_train_epochs=2`: Das Modell wird zweimal über das gesamte Trainings-Dataset iterieren.
*   `logging_steps=10`: Logs (wie Verlust und Lernrate) werden alle 10 Schritte ausgegeben.
*   `save_strategy="epoch"` und `save_total_limit=2`: Dies bedeutet, dass das Modell am Ende jeder Epoche gespeichert wird, aber nur die zwei neuesten Checkpoints behalten werden, um Speicherplatz zu sparen.
*   `bf16=True`: Aktiviert das `bfloat16`-Mixed-Precision-Training. Dies spart GPU-Speicher und kann die Trainingsgeschwindigkeit erhöhen, insbesondere auf GPUs, die bfloat16-Operationen nativ unterstützen (wie Tensor Cores auf NVIDIA A100 GPUs), ohne die Modellgenauigkeit wesentlich zu beeinträchtigen.
*   `report_to="none"`: Deaktiviert die Integration mit externen Reporting-Tools wie TensorBoard, Weights & Biases oder MLflow.
*   `optim="adamw_torch"`: Der verwendete Optimierungsalgorithmus ist `AdamW` (Adam mit Gewichtsverfall), implementiert von PyTorch. `AdamW` ist eine beliebte Wahl für das Training von Large Language Models.
*   `warmup_ratio=0.05` und `lr_scheduler_type="cosine"`: Dies konfiguriert den Lernraten-Scheduler. Zuerst wird die Lernrate über die ersten 5% der Trainingsschritte linear von 0 auf die angegebene `learning_rate` erhöht (Warmup), und danach folgt sie einem Cosinus-Schema, das die Lernrate allmählich reduziert.
*   `run_name`: Ein eindeutiger Name für den aktuellen Trainingslauf, der in den Logs und bei der Benennung von Verzeichnissen verwendet wird.

**4. Rolle des `Trainer`s:**
Der `Trainer` von Hugging Face ist eine leistungsstarke und benutzerfreundliche API, die den Trainingsprozess abstrahiert. Er wird initialisiert mit:
*   `model`: Das LoRA-Modell, das trainiert werden soll.
*   `args`: Die zuvor definierten `TrainingArguments`.
*   `train_dataset`: Das tokenisierte Trainings-Dataset (`tokenized_ds`).
*   `tokenizer`: Der `AutoTokenizer`, der verwendet wurde, um die Daten zu tokenisieren und für die Decodierung während des Trainings oder der Evaluation notwendig ist.
Der `Trainer` kümmert sich um die Trainingsschleife, Optimierungsschritte, Lernraten-Scheduling, Logging, Checkpoint-Speicherung und vieles mehr.

**5. Starten des Fine-Tunings (`trainer.train()`):**
Durch den Aufruf von `trainer.train()` wird der Fine-Tuning-Prozess gestartet. Der `Trainer` führt die Trainingsschleife gemäß den in `TrainingArguments` definierten Einstellungen aus.

**6. Speichern der LoRA-Adapter und des Tokenizers:**
Nachdem das Training abgeschlossen ist, werden die trainierten LoRA-Adapter und der Tokenizer gespeichert:
*   `model.save_pretrained(out_dir)`: Speichert die trainierbaren LoRA-Adaptergewichte und deren Konfiguration im angegebenen `out_dir`.
*   `tokenizer.save_pretrained(out_dir)`: Speichert die Tokenizer-Konfiguration und das Vokabular im selben `out_dir`. Dies ist entscheidend, um das fein abgestimmte Modell später mit dem korrekten Tokenizer laden und verwenden zu können.

*   `run_name`: Ein Bezeichner für den aktuellen Trainingslauf, nützlich für das Tracking, falls externe Tools wie Weights & Biases integriert wären.
*   `epochs`: Die Anzahl der Trainingsepochen, d.h. wie oft das gesamte Dataset während des Trainings durchlaufen wird.
*   `lr`: Die anfängliche Lernrate, die bestimmt, wie stark die Modellgewichte bei jedem Optimierungsschritt angepasst werden.
*   `batch_size`: Die Anzahl der Beispiele, die pro Trainingsschritt auf einem einzelnen Gerät verarbeitet werden. Hier auf 1 gesetzt, was bedeutet, dass jedes Beispiel einzeln verarbeitet wird.
*   `grad_acc`: Die Gradientenakkumulationsschritte. Dieser Parameter gibt an, wie viele Schritte die Gradienten akkumuliert werden, bevor ein Optimierungsschritt durchgeführt wird. Bei `batch_size=1` und `grad_acc=16` entspricht dies effektiv einer Batch-Größe von 16.

**2. Initialisierung des LoRA-Modells (`build_lora_model()`):**
Innerhalb von `train_lora` wird die `build_lora_model()`-Funktion aufgerufen. Diese Funktion ist verantwortlich für:
*   Das Laden des `BASE_MODEL` (`mistralai/Mistral-7B-Instruct-v0.2`) mithilfe von `AutoModelForCausalLM.from_pretrained()`. Dabei wird `torch.bfloat16` als `torch_dtype` verwendet, um das Modell im `bfloat16`-Format zu laden, was Speicher spart und die Trainingsgeschwindigkeit auf kompatibler Hardware erhöht. `device_map="auto"` sorgt dafür, dass das Modell automatisch auf die verfügbaren GPUs verteilt wird.
*   `model.config.use_cache = False` wird gesetzt, was für das Training wichtig ist, um Probleme mit dem Gradientenfluss zu vermeiden.
*   `model.gradient_checkpointing_enable()` wird aktiviert, um den Speicherverbrauch während des Trainings weiter zu reduzieren, indem die Aktivierungen bestimmter Schichten nicht im Speicher gehalten, sondern bei Bedarf neu berechnet werden.
*   Eine `LoraConfig` wird erstellt, die die Parameter für den LoRA-Adapter definiert (z.B. `r`, `lora_alpha`, `lora_dropout`, `bias`, `target_modules`).
*   Schließlich wird das Basismodell mit der `get_peft_model()`-Funktion um die LoRA-Adapter erweitert. Die Funktion gibt auch die Anzahl der trainierbaren Parameter des modifizierten Modells aus, was nützlich ist, um den Umfang des Fine-Tunings zu verstehen.

**3. `TrainingArguments` zur Konfiguration des Trainingsprozesses:**
Die `TrainingArguments`-Klasse wird verwendet, um alle Hyperparameter und Konfigurationen für den Trainingsprozess festzulegen:
*   `output_dir`: `/content/lora_det_final` oder `/content/lora_slot_final`. Dies ist der Pfad, unter dem alle Trainingsartefakte, einschließlich Modell-Checkpoints, Logs und der endgültigen LoRA-Adapter, gespeichert werden.
*   `per_device_train_batch_size=1`: Die Anzahl der Trainingsbeispiele, die pro GPU in einem einzigen Schritt verarbeitet werden. Ein Wert von 1 wird oft verwendet, wenn der Speicher sehr begrenzt ist oder wenn die effektive Batch-Größe durch Gradientenakkumulation erhöht wird.
*   `gradient_accumulation_steps=16`: Da die `per_device_train_batch_size` 1 ist, werden die Gradienten von 16 einzelnen Schritten gesammelt und dann einmal angewendet, was einer effektiven Batch-Größe von 16 entspricht. Dies hilft, die Stabilität des Trainings zu verbessern und größere effektive Batch-Größen zu simulieren, ohne den GPU-Speicher direkt zu überlasten.
*   `learning_rate=2e-4`: Die anfängliche Lernrate für den Optimierer. Dieser Wert wird im Laufe des Trainings oft durch einen Lernraten-Scheduler angepasst.
*   `num_train_epochs=2`: Das Modell wird zweimal über das gesamte Trainings-Dataset iterieren.
*   `logging_steps=10`: Logs (wie Verlust und Lernrate) werden alle 10 Schritte ausgegeben.
*   `save_strategy="epoch"` und `save_total_limit=2`: Dies bedeutet, dass das Modell am Ende jeder Epoche gespeichert wird, aber nur die zwei neuesten Checkpoints behalten werden, um Speicherplatz zu sparen.
*   `bf16=True`: Aktiviert das `bfloat16`-Mixed-Precision-Training. Dies spart GPU-Speicher und kann die Trainingsgeschwindigkeit erhöhen, insbesondere auf GPUs, die bfloat16-Operationen nativ unterstützen (wie Tensor Cores auf NVIDIA A100 GPUs), ohne die Modellgenauigkeit wesentlich zu beeinträchtigen.
*   `report_to="none"`: Deaktiviert die Integration mit externen Reporting-Tools wie TensorBoard, Weights & Biases oder MLflow.
*   `optim="adamw_torch"`: Der verwendete Optimierungsalgorithmus ist `AdamW` (Adam mit Gewichtsverfall), implementiert von PyTorch. `AdamW` ist eine beliebte Wahl für das Training von Large Language Models.
*   `warmup_ratio=0.05` und `lr_scheduler_type="cosine"`: Dies konfiguriert den Lernraten-Scheduler. Zuerst wird die Lernrate über die ersten 5% der Trainingsschritte linear von 0 auf die angegebene `learning_rate` erhöht (Warmup), und danach folgt sie einem Cosinus-Schema, das die Lernrate allmählich reduziert.
*   `run_name`: Ein eindeutiger Name für den aktuellen Trainingslauf, der in den Logs und bei der Benennung von Verzeichnissen verwendet wird.

**4. Rolle des `Trainer`s:**
Der `Trainer` von Hugging Face ist eine leistungsstarke und benutzerfreundliche API, die den Trainingsprozess abstrahiert. Er wird initialisiert mit:
*   `model`: Das LoRA-Modell, das trainiert werden soll.
*   `args`: Die zuvor definierten `TrainingArguments`.
*   `train_dataset`: Das tokenisierte Trainings-Dataset (`tokenized_ds`).
*   `tokenizer`: Der `AutoTokenizer`, der verwendet wurde, um die Daten zu tokenisieren und für die Decodierung während des Trainings oder der Evaluation notwendig ist.
Der `Trainer` kümmert sich um die Trainingsschleife, Optimierungsschritte, Lernraten-Scheduling, Logging, Checkpoint-Speicherung und vieles mehr.

**5. Starten des Fine-Tunings (`trainer.train()`):**
Durch den Aufruf von `trainer.train()` wird der Fine-Tuning-Prozess gestartet. Der `Trainer` führt die Trainingsschleife gemäß den in `TrainingArguments` definierten Einstellungen aus.

**6. Speichern der LoRA-Adapter und des Tokenizers:**
Nachdem das Training abgeschlossen ist, werden die trainierten LoRA-Adapter und der Tokenizer gespeichert:
*   `model.save_pretrained(out_dir)`: Speichert die trainierbaren LoRA-Adaptergewichte und deren Konfiguration im angegebenen `out_dir`.
*   `tokenizer.save_pretrained(out_dir)`: Speichert die Tokenizer-Konfiguration und das Vokabular im selben `out_dir`. Dies ist entscheidend, um das fein abgestimmte Modell später mit dem korrekten Tokenizer laden und verwenden zu können.

## 6. Inferenz und Adapter-Zusammenführung

Nachdem die LoRA-Adapter erfolgreich trainiert und gespeichert wurden, besteht der nächste Schritt darin, diese Adapter für die Textgenerierung zu nutzen. Dies beinhaltet das Laden des ursprünglichen Basismodells, das Hinzufügen der trainierten LoRA-Adapter und deren optionales Zusammenführen, um ein effizientes, feingetuntes Modell für die Inferenz zu erhalten. Abschließend werden die Generierungsparameter angepasst, um die gewünschte Ausgabe für jede spezifische Aufgabe zu steuern.

### 6.1. Tokenizer Initialisierung

Der `tokenizer`, der während des Trainings verwendet wurde, muss erneut geladen werden, um sicherzustellen, dass die Eingabetexte für die Inferenz korrekt tokenisiert und die generierten Ausgaben richtig dekodiert werden können. Der `AutoTokenizer` wird dazu mit dem `BASE_MODEL` initialisiert, und falls kein `pad_token` definiert ist, wird der `eos_token` als solcher zugewiesen, was für die Stapelverarbeitung (`batching`) während der Generierung entscheidend ist.

### 6.2. Parameter für die Textgenerierung

Die Qualität, Länge und der Stil des generierten Textes hängen maßgeblich von den gewählten Generierungsparametern ab. Hier sind die wichtigsten, die im Kontext dieses Projekts relevant sind:

*   **`max_new_tokens`**: Definiert die maximale Anzahl von *neuen* Tokens, die das Modell nach dem Prompt generieren soll. Die Generation stoppt, sobald diese Anzahl erreicht ist oder ein End-of-Sequence (EOS)-Token generiert wird. Dies ermöglicht eine direkte Kontrolle über die Länge der Antwort. Für die Detektionsaufgabe, die nur ein Token ("JA" oder "NEIN") erfordert, ist ein kleiner Wert wie `5` sinnvoll, um unnötige Generierung zu vermeiden. Für die Slot-Extraktion, die ein JSON-Objekt produziert, das länger sein kann, wird ein Wert wie `100` verwendet.

*   **`pad_token_id` (und `eos_token_id`)**: `pad_token_id` ist der Token, der verwendet wird, um kürzere Eingabesequenzen auf die maximale Länge einer Batch-Eingabe aufzufüllen. `eos_token_id` ist der Token, der das Ende einer generierten Sequenz signalisiert. Viele Modelle verwenden den `eos_token_id` auch als `pad_token_id`. `pad_token_id` ist wichtig für das Batching von Eingaben während der Generation. Das Modell stoppt die Generation normalerweise, wenn es seinen `eos_token_id` generiert. Ohne einen definierten `pad_token` kann es zu Fehlern beim Batching oder bei der Generierung kommen.

*   **`do_sample`**: Ein Boolescher Wert, der bestimmt, ob die Textgenerierung auf **Sampling** (zufällige Auswahl basierend auf Wahrscheinlichkeiten) oder **Greedy Decoding** (immer das wahrscheinlichste nächste Token wählen) basiert.
    *   `do_sample=False`: Greedy Decoding (deterministisch, kann zu repetitiven oder generischen Antworten führen).
    *   `do_sample=True`: Sampling (führt zu vielfältigeren und kreativeren Antworten, aber auch zu mehr Zufälligkeit).
    Bei `do_sample=True` werden weitere Parameter wie `temperature`, `top_k` und `top_p` aktiv, um das Sampling zu steuern. Wenn `do_sample=False`, werden diese Parameter ignoriert.

*   **`temperature` – Die Steuerung der Zufälligkeit**: Wird nur bei `do_sample=True` verwendet. Ein Wert größer als 0, der die Zufälligkeit der generierten Ausgabe steuert. `temperature` ist ein Parameter, der die **Wahrscheinlichkeitsverteilung** der möglichen nächsten Token steuert. Stellen Sie sich vor, das Modell schlägt für jedes nächste Wort eine Liste von Kandidaten vor, jeder mit einer bestimmten Wahrscheinlichkeit (z.B. "Katze" mit 80%, "Hund" mit 15%, "Haus" mit 4%, "Baum" mit 1%).
    *   **`temperature` = 1.0 (Standard)**: Bei einer Temperatur von 1.0 werden die Wahrscheinlichkeiten so verwendet, wie das Modell sie berechnet hat. Die Wahl des nächsten Tokens erfolgt proportional zu diesen Wahrscheinlichkeiten.
    *   **`temperature` < 1.0 (z.B. 0.7)**: Eine niedrigere Temperatur **schärft die Wahrscheinlichkeitsverteilung**. Das bedeutet, dass die Wahrscheinlichkeiten für die bereits wahrscheinlichen Tokens noch höher werden, während die unwahrscheinlicheren Tokens noch unwahrscheinlicher werden. Das Ergebnis ist eine **deterministischere und konservativere** Generierung. Das Modell wählt eher die "sicheren", häufigsten oder vorhersehbarsten Wörter. Dies ist nützlich für Aufgaben, die präzise, faktenbasierte oder weniger kreative Antworten erfordern.
    *   **`temperature` > 1.0 (z.B. 1.2)**: Eine höhere Temperatur **glättet die Wahrscheinlichkeitsverteilung**. Das führt dazu, dass die Wahrscheinlichkeiten zwischen allen möglichen Tokens "gerechter" verteilt werden – die Differenzen zwischen wahrscheinlichen und unwahrscheinlichen Tokens werden kleiner. Dies fördert die **Zufälligkeit und Vielfalt** in der Generierung. Das Modell kann "überraschendere" oder kreativere Wörter wählen, birgt aber auch das Risiko, weniger kohärenten oder sogar unsinnigen Text zu erzeugen.

*   **`top_k`**: Wird nur bei `do_sample=True` verwendet. Das Modell wählt das nächste Wort nur aus den `k` wahrscheinlichsten Wörtern aus. Dies begrenzt den Wortschatz für das Sampling auf die top-k wahrscheinlichsten Tokens und reduziert die Wahrscheinlichkeit von sehr unwahrscheinlichen Tokens.

*   **`top_p` (Nucleus Sampling) – Dynamische Wortschatzauswahl**: Wird nur bei `do_sample=True` verwendet. `top_p` bietet eine dynamischere und oft effektivere Methode zur Steuerung der Vielfalt als `top_k`. Anstatt eine feste Anzahl von `k` Tokens zu wählen, berücksichtigt `top_p` eine variable Anzahl von Tokens basierend auf deren kumulierten Wahrscheinlichkeit:
    *   **Mechanismus**: Bei `top_p=0.9` (ein häufig verwendeter Wert) werden alle Tokens von den höchstwahrscheinlichsten absteigend so lange ausgewählt, bis deren **kumulierte Wahrscheinlichkeit den Schwellenwert `p` überschreitet**. Aus dieser dynamisch erstellten Menge von Tokens wird dann zufällig das nächste Wort ausgewählt.
    *   **Vorteile gegenüber `top_k`**: Es passt sich der Verteilung der Wahrscheinlichkeiten an: Wenn die Verteilung spitz ist (einige wenige Tokens sind sehr wahrscheinlich), wählt `top_p` aus einer kleinen Menge; wenn die Verteilung flacher ist, wählt es aus einer größeren Menge. Dies kann zu natürlicheren und vielfältigeren Texten führen als ein fester `top_k` Wert.

*   **`num_beams` (für Beam Search)**: Wird nur bei `do_sample=False` verwendet. Aktiviert die Beam Search-Generierungsstrategie. Anstatt nur den wahrscheinlichsten Pfad zu verfolgen (wie bei Greedy Decoding), hält Beam Search `num_beams` viele vielversprechende Kandidatensequenzen gleichzeitig im Auge. Führt zu qualitativ besseren, kohärenteren Texten, ist aber rechenintensiver.

### 6.3. LoRA-Adapter für Detektion: Laden und Inferenz

Für die normative Detektionsaufgabe wird das `Mistral-7B-Instruct-v0.2` Basismodell geladen, wobei `torch.bfloat16` für Speichereffizienz und `device_map="auto"` für die optimale GPU-Nutzung verwendet wird (analog zur Funktionalität in Zelle `565d680e`). Anschließend werden die zuvor trainierten LoRA-Adapter aus dem Verzeichnis `/content/lora_det_final` mithilfe von `PeftModel.from_pretrained` geladen und mit dem Basismodell zusammengeführt (`merge_and_unload`). Dieser Schritt erzeugt ein eigenständiges, feingetuntes Modell (`merged_model_det`), das direkt für die Inferenz verwendet werden kann, ohne das ursprüngliche Basismodell separat halten zu müssen.

Die Inferenz erfolgt über die `generate_response`-Funktion. Diese Funktion tokenisiert den gegebenen Prompt, leitet ihn durch das `merged_model_det` und dekodiert die generierten Token zurück in Text. Für die Detektionsaufgabe ist es entscheidend, `max_new_tokens=5` zu setzen, da die erwartete Antwort lediglich "JA" oder "NEIN" ist, wie in Zelle `68464dd7` demonstriert. Dies stellt sicher, dass das Modell keine unnötig langen oder irrelevanten Textabschnitte generiert.

### 6.4. LoRA-Adapter für Slot-Extraktion: Laden und Inferenz

Ähnlich wie bei der Detektion wird für die Slot-Extraktionsaufgabe das Basismodell erneut geladen. Die LoRA-Adapter für die Slot-Extraktion werden aus `/content/lora_slot_final` geladen und mit diesem Basismodell zusammengeführt, um das `merged_model_slot` zu erhalten (analog zur Funktionalität in Zelle `40ddfee5`). Dies ermöglicht eine spezialisierte Inferenz für die Extraktion von PARIS-Slots.

Die Inferenz für die Slot-Extraktion verwendet ebenfalls die `generate_response`-Funktion. Da die Ausgabe ein strukturiertes JSON-Objekt sein soll, das typischerweise länger als ein einzelnes Wort ist, wird `max_new_tokens=100` gesetzt, um sicherzustellen, dass das Modell genügend Raum hat, die vollständige JSON-Antwort zu generieren, wie in Zelle `1e221fdf` gezeigt. Die dekodierte Antwort wird dann zurückgegeben und kann weiterverarbeitet werden.
